```python
# Load packages
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np; np.random.seed(1)
from sklearn import datasets
import pandas as pd
import functools as ft
import seaborn as sns
import re  
from scipy.stats import logistic
import calendar
import random
```

```python
# Load data
youtube_data = pd.read_csv('US_youtube_trending_data.csv')
lookup = pd.read_csv('category_ids.csv')
clean_data_20 = pd.read_csv('clean_data_20.csv')
```

```python
#see if subscribercount could impact the amount of like/dislike a channel gets
#see if number of video a channel produces could impact the amount of channel view count 
fig, (ax) = plt.subplots(figsize=(12, 6))
#format
plt.style.use('ggplot')
#supplot 1
#initialize
df = clean_data_20.groupby('channelTitle').sum() 

#subplot (rows, columns, locate)
ax = plt.subplot(1, 2, 1)
ax.scatter(df['videoCount'], df['likeCount'], color='y')
ax.scatter(df['videoCount'], df['dislikeCount'], color='m')
ax.scatter(df['videoCount'], df['commentCount'], color='steelblue')
ax.set(title='Title',  xlabel='Video Count', ylabel='Reactions')
ax.legend(['likes', 'dislikes', 'comments'])
plt.setp(ax.xaxis.get_majorticklabels(), rotation=60)


#format
plt.style.use('ggplot')
#initialize, group by date
df = clean_data_20.groupby('channelTitle').sum() 
#subplot
ax = plt.subplot(1, 2, 2) #subplot (rows, columns, locate)
ax.scatter(df['subscriberCount'], df['likeCount'], color='y') 
ax.scatter(df['subscriberCount'], df['dislikeCount'], color='m')  
ax.scatter(df['subscriberCount'], df['commentCount'], color='steelblue')  
#labels
ax.set(title='Title', xlabel='Subscriber Count', ylabel='Reactions')
ax.legend(['likes', 'dislikes', 'comments'])
plt.setp(ax.xaxis.get_majorticklabels(), rotation=60)
plt.show()
```
